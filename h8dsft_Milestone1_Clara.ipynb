# Identifikasi Masalah

## Latar belakang

Pada era digital saat ini, profesi di bidang ilmu data telah menjadi salah satu pekerjaan yang paling diminati.` Permintaan` akan `data scientist` meningkat seiring dengan kemajuan teknologi dan meningkatnya kesadaran perusahaan akan pentingnya data dalam pengambilan keputusan.

## Topik Permasalahan

Topik utama dalam penelitian ini adalah untuk `menganalisis tren pekerjaan di data scientist`. Penelitian ini bertujuan untuk memahami berbagai faktor yang mempengaruhi gaji, distribusi pekerjaan berdasarkan lokasi, dan keterampilan yang paling dibutuhkan oleh perusahaan.

## Problem Statement

Penelitian ini bertujuan untuk mengidentifikasi distribusi gaji, keterampilan yang dibutuhkan, distribusi pekerjaan berdasarkan lokasi, dan pengaruh rating perusahaan terhadap penawaran gaji serta peluang karir di bidang data scientist pada tahun 2021. Dengan menggunakan metode SMART, penelitian ini akan mengukur distribusi gaji dan keterampilan yang dibutuhkan, serta jumlah pekerjaan berdasarkan lokasi, dan menghitung korelasi antara rating perusahaan dan penawaran gaji.

## Metode SMART  



**S (Specific)**  
Mengidentifikasi distribusi gaji, keterampilan yang dibutuhkan, distribusi pekerjaan berdasarkan lokasi, dan pengaruh rating perusahaan terhadap penawaran gaji dan peluang karir di bidang Data Scientst pada tahun 2021.

**M (Measurable)**  
Mengukur distribusi gaji dan keterampilan yang dibutuhkan, serta jumlah pekerjaan berdasarkan lokasi. Menghitung korelasi antara rating perusahaan dan penawaran gaji.

**A (Achievable)**  
Menggunakan dataset yang ada untuk melakukan analisis statistik, membuat visualisasi data, dan menghitung korelasi. Proses ini melibatkan pengumpulan data, pembersihan data, analisis deskriptif, dan analisis inferensial.

**R (Relevant)**  
Penelitian ini relevan untuk memahami tren pekerjaan di bidang Data Scientst, membantu perusahaan dalam pengambilan keputusan rekrutmen, dan memberikan wawasan bagi pencari kerja tentang keterampilan yang dibutuhkan dan lokasi dengan peluang pekerjaan terbanyak.

**T (Time-bound)**  
Penelitian ini menggunakan data dari tahun 2021 dan diharapkan dapat membantu menyelesaikan permasalah pada awal tahun 2025 untuk memberikan wawasan terkini bagi keputusan yang akan datang.

## 1. Negara bagian mana yang memiliki jumlah pekerjaan terbanyak?

`Statistik Deskriptif`: Menampilkan distribusi jumlah lowongan pekerjaan di 10 negara bagian dalam bentuk tabel dan grafik batang

```python
# Menambahkan label untuk negara bagian
lab = ["California", "Massachusetts", "New York", "Virginia", "Illinois", "Maryland", "Pennsylvania", "Texas", "Washington", "North Carolina"]

# Mendapatkan 10 negara bagian teratas dengan jumlah lowongan pekerjaan terbanyak
top_states = df["Job Location"].value_counts().nlargest(10)

# Mengatur ukuran gambar
plt.rcParams['figure.figsize'] = 12, 5

# Membuat plot batang
fig, ax = plt.subplots()  # Membuat subplot
sns.barplot(x=top_states.index, y=top_states.values, ax=ax)  # Membuat plot batang dengan sumbu-x dan sumbu-y

# Menghapus border atas dan kanan
sns.despine(bottom=False, left=False)

# Menambahkan persentase pada setiap batang
for p in ax.patches:
    percentage = '{:.2f}%'.format((p.get_height() / len(df)) * 100)  # Menghitung persentase
    ax.annotate(percentage, (p.get_x() + 0.1, p.get_height() + 4.5), fontsize=10)  # Menambahkan teks persentase di atas setiap batang

# Mempercantik plot
plt.title('\nJumlah Perkerjaan terbanyak di negara bagian\n', size=16, color='black')  # Menambahkan judul plot
plt.xticks(fontsize=13)  # Mengatur ukuran font pada label sumbu-x
plt.yticks(fontsize=12)
plt.xlabel('\nNegara Bagian \n', fontsize=13, color='black')
plt.ylabel('\nCount\n', fontsize=13, color='black')

# Menambahkan legenda dengan label negara bagian
patches = [matplotlib.patches.Patch(color=sns.color_palette()[i], label=j) for i, j in zip(range(0, 10), lab)]
plt.legend(handles=patches, loc="upper right")

# Menampilkan plot
plt.show()
```

### Kesimpulan

Dari grafik di atas, kita dapat menyimpulkan bahwa:

1. **California (CA)** memiliki jumlah lowongan pekerjaan terbanyak dengan persentase 20.49%.
2. **Massachusetts (MA)** berada di posisi kedua dengan 13.88% dari total lowongan pekerjaan.
3. **New York (NY)** menempati posisi ketiga dengan 9.70% lowongan pekerjaan.
4. Negara bagian lainnya seperti **Virginia (VA)**, **Illinois (IL)**, **Maryland (MD)**, **Pennsylvania (PA)**, **Texas (TX)**, **North Carolina (NC)**, dan **Washington (WA)** juga memiliki kontribusi yang signifikan terhadap jumlah lowongan pekerjaan, meskipun dengan persentase yang lebih rendah.

## 3. Berapa rata-rata gaji minimum dan maksimum di berbagai negara, dan negara mana yang menawarkan gaji tertinggi?

Menghitung Frekuensi dan `Mengurutkan "Job Location"`

```python

# Menghitung frekuensi masing-masing "Job Location" dan mengurutkannya
sorter = df["Job Location"].value_counts().index

# Mengubah urutan kategori dalam kolom "Job Location"
df["Job Location"] = df["Job Location"].astype("category")
df["Job Location"] = df["Job Location"].cat.set_categories(sorter)

# Mengurutkan DataFrame dan mengatur ulang indeks
df = df.sort_values(["Job Location"]).reset_index(drop=True)

# Menampilkan hasil
print(df.head())
```

Membuat Plot untuk menampilkan negara 

```python

# Menambahkan label untuk negara bagian
lab = []
for i in sorter[0:10]:
    lab.append(i)  # Menambahkan nama negara bagian ke dalam daftar lab berdasarkan urutan dalam sorter

# Mengatur lokasi label dan lebar batang
x = np.arange(len(lab))  # lokasi label pada sumbu x
width = 0.35  # lebar batang

# Membuat plot batang
fig, ax = plt.subplots(1)  # Membuat figure dan axes untuk plot
rects1 = ax.bar(x - width/2, df["Lower Salary"][0:10], width, label='Gaji Minimal')  # Membuat batang untuk gaji minimal
rects2 = ax.bar(x + width/2, df["Upper Salary"][0:10], width, label='Gaji Maksimal')  # Membuat batang untuk gaji maksimal

# Mempercantik plot
plt.title('\nNegara Bagian dengan Gaji Minimal dan Maksimal Rata-Rata Tahunan\n', size=16, color='black')  # Menambahkan judul plot
plt.xticks(fontsize=13)  # Mengatur ukuran font untuk label sumbu x
plt.yticks(fontsize=12)  # Mengatur ukuran font untuk label sumbu y
plt.xlabel('\nNegara Bagian\n', fontsize=13, color='black')  # Menambahkan label sumbu x
plt.ylabel('\nGaji (K)\n', fontsize=13, color='black')  # Menambahkan label sumbu y

# Mengatur label sumbu x dan menambahkan legenda
ax.set_xticks(x)  # Mengatur posisi  pada sumbu x
ax.set_xticklabels(lab)  # Mengatur label untuk pada sumbu x
ax.legend()  # Menambahkan legenda untuk plot

# Menampilkan plot
plt.show()  # Menampilkan plot yang telah dibuat
```

### Negara Bagian dengan Gaji Tertinggi

Berdasarkan grafik, negara bagian yang menawarkan gaji maksimal tertinggi adalah:

1. **California (CA)** dengan gaji maksimal sekitar $150K.
2. **Illinois (IL)** dengan gaji maksimal sekitar $140K.
3. **Massachusetts (MA)** dengan gaji maksimal sekitar $130K (sama dengan Maryland, Virginia, North Carolina, dan Washington).

## 4. Perusahaan mana yang menawarkan gaji tertinggi

`Statistik Deskriptif`: Menampilkan rata-rata gaji tertinggi yang ditawarkan oleh perusahaan dalam bentuk tabel dan grafik.

```python
# Menghitung rata-rata gaji per perusahaan
df['Avg Salary'] = (df['Lower Salary'] + df['Upper Salary']) / 2
rata_rata_gaji_per_perusahaan = df.groupby('Company Name')['Avg Salary'].mean().sort_values(ascending=False)

# Menampilkan perusahaan-perusahaan dengan rata-rata gaji tertinggi
tabel_gaji_tertinggi = rata_rata_gaji_per_perusahaan.head(10).reset_index()
tabel_gaji_tertinggi.columns = ['Nama Perusahaan', 'Rata-rata Gaji']

# Membuat plot perusahaan-perusahaan dengan rata-rata gaji tertinggi
plt.figure(figsize=(10, 6))
sns.barplot(x='Nama Perusahaan', y='Rata-rata Gaji', data=tabel_gaji_tertinggi)
plt.xticks(rotation=90)  # Memutar label sumbu X agar lebih mudah dibaca
plt.xlabel('Nama Perusahaan')  # Memberi label pada sumbu X
plt.ylabel('Rata-rata Gaji (dalam ribuan)')  # Memberi label pada sumbu Y
plt.title('10 Perusahaan dengan Rata-rata Gaji Tertinggi di Data Science')  # Judul plot
plt.show()  # Menampilkan plot
```

`Perusahaan yang menawarkan gaji tertinggi` di bidang data science, berdasarkan gambar yang menunjukkan rata-rata gaji, adalah `Gallup.`

- Menampilkan hasil uji ANOVA, termasuk nilai F-statistic dan p-value.

```python
# Perform ANOVA to compare average minimum salaries across locations
anova_min_result = f_oneway(*[group['Lower Salary'].values for name, group in df.groupby('Location')])

# Perform ANOVA to compare average maximum salaries across locations
anova_max_result = f_oneway(*[group['Upper Salary'].values for name, group in df.groupby('Location')])

# Display ANOVA results
anova_results = {
    "Min Salary ANOVA": {
        "F-statistic": anova_min_result.statistic,
        "p-value": anova_min_result.pvalue
    },
    "Max Salary ANOVA": {
        "F-statistic": anova_max_result.statistic,
        "p-value": anova_max_result.pvalue
    }
}
anova_results
```



#### Hasil Uji ANOVA untuk Gaji Minimum

- **p-value**: 1.06e-20

#### Hasil Uji ANOVA untuk Gaji Maksimum

- **p-value**: 2.05e-24

**Interpretasi**:
- Nilai p-value yang sangat kecil (jauh di bawah 0.05) menunjukkan bahwa terdapat perbedaan yang signifikan dalam rata-rata gaji minimum dan maksimum di berbagai lokasi.

## 8.  Menganalisis hubungan antara rating perusahaan dengan gaji yang ditawarkan

**Statistik Deskriptif**

1. Menghitung Rata-rata Gaji:
 - Memuat dataset yang mengandung informasi tentang rating perusahaan dan gaji.
 - Menghitung rata-rata gaji (Lower Salary, Upper Salary, atau kombinasi keduanya) untuk setiap perusahaan.
 - Menggabungkan data rating perusahaan dengan data rata-rata gaji.
- Menghasilkan tabel dengan rata-rata rating dan gaji untuk setiap perusahaan.

```python
# Menghitung rata-rata gaji (Anda bisa menggunakan Gaji Bawah, Gaji Atas, atau kombinasi keduanya)
df['Avg Salary'] = (df['Lower Salary'] + df['Upper Salary']) / 2
# Menambahkan kolom baru 'Avg Salary' dengan menghitung rata-rata dari 'Lower Salary' dan 'Upper Salary'

# Mengelompokkan data berdasarkan Nama Perusahaan dan menghitung rata-rata rating dan gaji rata-rata
company_rating_salary = df.groupby('Company Name').agg({'Rating': 'mean', 'Avg Salary': 'mean'}).reset_index()
# Menggunakan groupby untuk mengelompokkan data berdasarkan 'Company Name'
# Menggunakan agg untuk menghitung rata-rata 'Rating' dan 'Avg Salary' untuk setiap perusahaan
# Menggunakan reset_index untuk mengatur ulang indeks sehingga hasilnya menjadi DataFrame biasa

# Menghapus baris yang memiliki nilai kosong di kolom 'Rating' atau 'Avg Salary'
company_rating_salary = company_rating_salary.dropna(subset=['Rating', 'Avg Salary'])
# Menggunakan dropna untuk menghapus baris yang memiliki nilai NaN (kosong) di kolom 'Rating' atau 'Avg Salary'

# Menampilkan 5 baris pertama dari DataFrame hasil
company_rating_salary.head()
# Menggunakan head() untuk menampilkan 5 baris pertama dari DataFrame hasil

```

**Statistik Inferensial**
1. Scatter Plot:

- Membuat scatter plot untuk memvisualisasikan hubungan antara rating perusahaan dan gaji yang ditawarkan.
- Scatter plot membantu untuk melihat pola atau tren dalam data.


2. Menghitung Korelasi:

- Menghitung nilai korelasi antara rating perusahaan dan rata-rata gaji untuk memahami seberapa kuat hubungan antara kedua variabel tersebut.
- Korelasi memberikan indikasi tentang apakah ada hubungan linear antara rating perusahaan dan gaji yang ditawarka

```python
# Membuat scatter plot
plt.figure(figsize=(12, 8))
# Mengatur ukuran plot menjadi 12x8 inci
sns.scatterplot(x='Rating', y='Avg Salary', data=company_rating_salary)
# Membuat scatter plot menggunakan seaborn dengan sumbu X adalah 'Rating' dan sumbu Y adalah 'Avg Salary'
plt.xlabel('Company Rating')
# Memberi label pada sumbu X
plt.ylabel('Average Salary')
# Memberi label pada sumbu Y
plt.title('Hubungan antara Rating Perusahaan dan Gaji Rata-rata')
# Memberi judul pada plot
plt.show()
# Menampilkan plot

# Menghitung korelasi
correlation = company_rating_salary['Rating'].corr(company_rating_salary['Avg Salary'])
# Menghitung koefisien korelasi antara 'Rating' dan 'Avg Salary' menggunakan fungsi corr
print(f'Korelasi antara rating perusahaan dan gaji rata-rata: {correlation}')
# Mencetak hasil korelasi

```


- **Rata-rata Gaji**: Perusahaan dengan rating lebih tinggi cenderung menawarkan gaji lebih tinggi.

**Interpretasi**:
